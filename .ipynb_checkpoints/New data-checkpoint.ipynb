```python
# Some tracks have unrecognized characters so we skip on lines that have them
tracks = pd.read_csv('Data\\tracks.tsv', sep='\t', on_bad_lines='skip')
```

```python
tracks = tracks[["track_id", "artist_name"]]
```

```python
users = pd.read_csv('Data\\users.tsv', sep='\t')
```

```python
# Filter users that have country data
Cusers = users[users["country"].notna()]
```

```python
# Get all unique user id's 
UCusers = Cusers["user_id"].unique()
```

```python
# Filter out tracks that don't have track information
Tlistcount = listcount[listcount["track_id"].isin(tracks["track_id"].unique())]
```

```python
# Filter out listening data of users that have no country information
Clistcount = Tlistcount[Tlistcount["user_id"].isin(UCusers)]
```

```python
# Dictionaries to convert ISO-alpha2 codes to user ids and vice versa
countries = Cusers["country"].unique()
countryUserIDs = {}
userIDsCountry = {}
for country in countries:
    IDs = Cusers[Cusers["country"] == country]["user_id"]
    countryUserIDs[country] = IDs
    for i in IDs:
        userIDsCountry[i] = country
```

```python
# Previous cells takes a long time to do, so store dictionaries and load them in here
with open('countryID.pkl', 'rb') as f:
    countryID = pickle.load(f)

with open('idCountry.pkl', 'rb') as f:
    idCountry = pickle.load(f)
    
with open('codesCountry.pkl', 'rb') as f:
    codesCountry = pickle.load(f)

with open('countryCodes.pkl', 'rb') as f:
    countryCodes = pickle.load(f)

with open('artistcol.pkl', 'rb') as f:
    artistcol = pickle.load(f)
    
with open('artistCountryCodes.pkl', 'rb') as f:
    artistCountryCodes = pickle.load(f)
    
with open('countryArtistID.pkl', 'rb') as f:
    countryArtistID = pickle.load(f)
```

### Country artist unique listeners

```python
Clistcount["artist"] = artistcol
```

```python
# Gives a dataframe with all unqiue user_id and artist combinations
CListCountUA = Clistcount.groupby(['user_id','artist']).size().reset_index().rename(columns={0:'count'})
```

```python
countryArtistsUL = np.array([[0] * len(artists)] * len(countries))
```

```python
countryListeningUL = np.array([[0] * len(countries)] * len(countries))
```

```python
i = 0
num = 100_000
for index, row in CListCountUA.iterrows():
    if i % num == 0:
        print(i, "/", len(CListCountUA), end="\r")
    userCountry = countryID[userIDsCountry[row["user_id"]]]
    artist = row["artist"]
    artID = artistID[artist]
    countryArtistsUL[userCountry][artID] += 1 # Add one for every unique combination
    # If the artist has country information add that to it's own array
    if artist in artistCountryCodes.keys():
        artistCode = artistCountryCodes[artist]
        if artistCode in countryID.keys():
            artistCountry = countryID[artistCode]
            countryListeningUL[userCountry][artistCountry] += 1
    i += 1
```

```python
with open('countryArtistsUL.pkl', 'wb+') as f:
    pickle.dump(countryArtistsUL, f)

with open('countryListeningUL.pkl', 'wb+') as f:
    pickle.dump(countryListeningUL, f)
```

```python
with open('countryArtistsUL.pkl', 'rb') as f:
    countryArtistsUL = pickle.load(f)

with open('countryListeningUL.pkl', 'rb') as f:
    countryListeningUL = pickle.load(f)
```

#### View result


```python
i = 0
for countryArtist in countryArtistsUL:
    print(codesCountry[idCountry[i]])
    sortArtist = np.argsort(countryArtist)[::-1]
    print("Topartists:")
    for j in range(0,10):
        print(str(j + 1) + '.', "Name:", idArtist[sortArtist[j]], ", Count:", countryArtist[sortArtist[j]])
    print("-------------------------------------")
    i += 1
```

```python
i = 0
for country in countryListeningUL:
    print(codesCountry[idCountry[i]])
    sortCountry = np.argsort(country)
    print("Top countries:")
    for j in range(1,11):
        print(str(j) + '.', "Country:", codesCountry[idCountry[sortCountry[-j]]], ", Listeners:", countryListeningUL[i][sortCountry[-j]])
        cArtists = countryArtistID[idCountry[sortCountry[-j]]]
        sortArtists = np.argsort(countryArtistsUL[i])
        k = 1
        l = 1
        while k <= 3:
            if sortArtists[-l] in cArtists:
                print("\t" + str(k) + '.', "Name:", idArtist[sortArtists[-l]], ", Listeners:", countryArtistsUL[i][sortArtists[-l]])
                k += 1
            l += 1
    print("-------------------------------------")
    i += 1
```

### Country country unique listeners

```python
countrycol = []
for artist in artistcol:
    if artist in artistCountryCodes.keys():
        countrycol.append(artistCountryCodes[artist])
    else:
        countrycol.append("None")
```

```python
Clistcount["country"] = countrycol
```

```python
# Gives a dataframe with all unqiue user_id and country combinations
CListCountUC = Clistcount.groupby(['user_id','country']).size().reset_index().rename(columns={0:'count'})
```

```python
CListCountUC = CListCountUC[CListCountUC["country"] != "None"]
```

```python
countryListeningUC = np.array([[0] * len(countries)] * len(countries))
```

```python
i = 0
num = 100_000
for index, row in CListCountUC.iterrows():
    if i % num == 0:
        print(i, "/", len(CListCountUC), end="\r")
    userCountry = countryID[userIDsCountry[row["user_id"]]]
    country = row['country']
    if country in countryID.keys():    
        cid = countryID[country]
        countryListeningUC[userCountry][cid] += 1
    i += 1
```

```python
with open('countryListeningUC.pkl', 'wb+') as f:
    pickle.dump(countryListeningUC, f)
```

#### View results

### Country artist count and unique tracks/listeners

#### Get the data

```python
Clistcount["artist"] = artistcol
```

```python
#Make array for data, rows is country ids and columns are artist ids
countryArtistsU = [[0] * len(artists)] * len(countries)
countryArtistsU = np.array(countryArtistsU)
```

```python
countryArtists = [[0] * len(artists)] * len(countries)
countryArtists = np.array(countryArtists)
```

```python
countrylistening = np.array([[0] * len(countries)] * len(countries))
countrylisteningU = np.array([[0] * len(countries)] * len(countries))
```

```python
i = 0
num = 100_000
for index, row in Clistcount.iterrows():
    if i % num == 0:
        print(i, "/", len(Clistcount), end="\r")
    country = countryID[userIDsCountry[row["user_id"]]]
    artist = row["artist"]
    aid = artistID[artist]
    countryArtists[country][aid] += row["count"] # Add count to get actual listening counts
    countryArtistsU[country][aid] += 1 # Add one for every unique combination 
    if artist in artistCountryCodes.keys():
        artistCode = artistCountryCodes[artist]
        if artistCode in countryID.keys():
            artistCountry = countryID[artistCode]
            countrylistening[country][artistCountry] += row["count"]
            countrylisteningU[country][artistCountry] += 1
            
    i += 1
```

#### View some initial result

```python
i = 0
for countryArtist in countryArtists:
    print(codesCountry[idCountry[i]])
    sortArtist = np.argsort(countryArtist)[::-1]
    print("Topartists:")
    for j in range(0,10):
        print(str(j + 1) + '.', "Name:", idArtist[sortArtist[j]], ", Count:", countryArtist[sortArtist[j]], ", Unique:", countryArtistsU[i][sortArtist[j]])
    print("-------------------------------------")
    i += 1
```

```python
i = 0
for countryArtist in countryArtistsU:
    print(codesCountry[idCountry[i]])
    sortArtist = np.argsort(countryArtist)
    print("Topartists:")
    for j in range(1,11):
        print(str(j) + '.', "Name:", idArtist[sortArtist[-j]], ", Unique:", countryArtist[sortArtist[-j]], ", Count:", countryArtists[i][sortArtist[-j]])
    print("-------------------------------------")
    i += 1
```

```python
aggArtists = countryArtists.sum(axis=0)
aggArtistsU = countryArtistsU.sum(axis=0)
aggSortArtists = np.argsort(aggArtists)
print("Topartists:")
for j in range(1,100000):
    print(str(j) + '.', "Name:", idArtist[aggSortArtists[-j]], ", Count:", aggArtists[aggSortArtists[-j]], ", Unique:", aggArtistsU[aggSortArtists[-j]]) 
```
